A linguagem Python fornece um conjunto de operadores artiméticos, que inclui os operadores binários para as quatro operações matemáticas básicas entre outros, tais como: <br>
<br>
**Operadores Unários** – Aceitam só um operando. Em Python, eles são -, +, ~, not (negação unária, positivo unário, negação bitwise e negação lógica). Observe que a linguagem Python não tem os operadores unários de adição ++ e subtração -- encontrados em outras linguagens, como C e Java.

**Operadores Binários** – Utilizam dois operandos para formar uma expressão.

**Operadores Ternários** – Também conhecidos como expressões condicionais ou “if ternário python“, precisam de dois operandos e uma expressão. Têm a forma:  `<operando1> if <expressão> else <operando2>`.

Vamos agora usar o NumPy para resolver o seguinte sistema de equações lineares: 

$$\left\{\begin{array} 22x + 5y - 10z = 99 \\ 7x - 13y + 4z = 118 \\ -3x + 4y - 2z = -45 \end{array} \right.$$

Podemos transformar esse sistema de equações lineares em um sistema matricial da forma:

$$Ax = B$$

Onde:

$$ A = \left[ \begin{array}{c} 2 & 5 & -10 \\ 7 & -13 & 4 \\ -3 & 4 & -2 \end{array} \right]$$

$$B = \left[ \begin{array}{c} -99 \\ 118 \\ -45 \end{array} \right]$$

$$x = \left[\begin{array}{c} x \\ y \\ z \end{array} \right]$$

Deste modo, a solução do sistema de equações será:

$$x = A^{-1}B$$

````python
A = np.array([[2,5,-10],[7,-13,4],[-3,4,-2]])
B = np.array([[-99],[118],[-45]])
````

````python
Ainv = np.linalg.inv(A)
Ainv
````

````python
x = np.matmul(Ainv,B)
x
````

````python
matmul(A,x)
````

## Matplotlib

Outra biblioteca muito útil para a engenharia é a biblioteca matplotlib. Ela é utilizada para a plotagem de gráficos.

Vamos plotar um gráfico da equação $y = x^3 - 5x^2 + 8x - 15$. Para isso utilizaremos os coneitos de funções e a biblioteca numpy. 

Primeiramente, vamos definir uma função para expressar essa biblioteca:

````python
def f(x):
    return x**3 - 5*x**3 + 8*x - 15
````

Como se trata de uma função simples, também podemos criá-la usando uma função anônima:

````python
f = lambda x: x**3 - 5*x**3 + 8*x - 15
````

Vamos criar um array para ser o eixo $x$ do nosso gráfico. Para isso precisamos criar uma array linearmente espaçado. O comando ```linspace``` faz exatamente isso:

````python
x = np.linspace(-20,20,2000) # criamos um array que vai de -20 a 20 com 2000 pontos
````

````python
y = f(x)
````

Vamos agora importar a biblioteca de plotagem de gráficos:

````python
import matplotlib.pyplot as plt
````

````python
plt.figure() # cria a figura
plt.plot(x,y) # plota o gráfico
plt.xlabel('x') # rótulo para o eixo x
plt.ylabel(r'$y = x^3 - 5x^2 + 8x - 15$') # rótulo para o eixo y
plt.title('Gráfico 1') # título do gráfico
plt.grid(True) # coloca uma grade no gráfico
plt.show()
````

O gráfico ficou pequeno, podemos aumentar o tamanho dele. Vamos aproveitar e mudar a cor da curva.

````python
plt.figure(figsize=(15,7)) # cria a figura
plt.plot(x,y,'red') # plota o gráfico
plt.xlabel('x') # rótulo para o eixo x
plt.ylabel(r'$y = x^3 - 5x^2 + 8x - 15$') # rótulo para o eixo y
plt.title('Gráfico 1') # título do gráfico
plt.grid(True) # coloca uma grade no gráfico
plt.show()
````

Podemos alterar a nítides do gráfico:

Vamos agora resolver o seguinte problema com os conhecimentos adquiridos:

No instante $t=0$ uma partícula de massa $m$ é liberada a partir do repouso de uma altura $h$ e cai verticalmente. Admitindo que a força de resistência do ar seja proporcional à velocidade instantânea, de acordo com a relação $f_r = -cv$, pede-se: **a)** desenvolver as funções descrevendo a posição, a velocidade e a aceleração da partícula em função do tempo transcorrido após sua liberação; **b)** traçar as curvas representando as funções obtidas no item a), adotando $m=1.0kg$, $h=100.0 m$, $c=0.8Ns/m$.<br>
Solução:
$$-cv - mg = m \dfrac{dv}{dt}$$

````python
m,g,h = symbols('m g h')
c = symbols('c',nonzero=True,positive=True)
t = symbols('t')
v = Function('v')(t)
````

Reorganizando a equação de movimento:
$$m \dfrac{dv}{dt} +cv + mg = 0$$

````python
EqMov = Eq(m*v.diff(t) + c*v + m*g,0)
EqMov
````

````python
v_sol = dsolve(EqMov.lhs,v)
v_sol
````

````python
v_sol.subs(t,0)
````

````python
C1 = solve(v_sol.subs(t,0).rhs,'C1')[0]
C1
````

````python
v_sol = v_sol.subs('C1',C1).simplify()
v_sol
````

````python
a = v_sol.rhs.diff(t)
a
````

````python
x = integrate(v_sol.rhs,t)
x
````

````python
x += h
x
````

````python
valores = {m:1,h:100,c:0.8,g:9.8}
````

````python
v_num = v_sol.subs(valores).rhs
v_num
````

Vamos vericar a velocidade máxima aplicando um limite fazendo o tempo tender a infinito

````python
limit(v_num,t,oo)
````

````python
x_num = x.subs(valores)
x_num
````

````python
a_num = a.subs(valores)
a_num
````

````python
lam_x = lambdify(t,x_num,modules = ['numpy'])
t_vals = linspace(0,5,100)
x_vals = lam_x(t_vals)
plt.figure(dpi=150)
plt.plot(t_vals,x_vals)
plt.xlabel('Tempo [s]')
plt.ylabel('Posição [m]')
plt.title('Posição pelo tempo')
plt.grid(True)
````
